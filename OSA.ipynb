```python
import pandas as pd

# Import Sales Transaction file and Customer file
sales_2017_df = pd.read_csv('Sales Transactions-2017.csv')
sales_2018_df = pd.read_csv('Sales Transactions-2018.csv')
sales_2019_df = pd.read_csv('Sales Transactions-2019.csv')
```

Perform Data Cleaning 


```python
# Concatenate all the Sales dataframes
full_sales_df = pd.concat([sales_2017_df, sales_2018_df, sales_2019_df], ignore_index=True,sort=False)
```

```python
# Drop the Columns Gross, Disc, Voucher Amount, using drop() method on the specific columns
sales_df = full_sales_df.drop(columns=['Gross','Disc','Voucher Amount'],axis=1)
```

```python
sales_df.head(15)
```

```python
# Dropping the Rows with Date column being NaN (Null) or Spaces
sales_df = sales_df[sales_df['Date'] != ' '].dropna(subset=['Date'])
sales_df.head(15)
```

```python
# Convert the Date format YYYY/MM/DD 
sales_df['Date'] = pd.to_datetime(sales_df['Date'],dayfirst=True)
```

```python
# Removing the string
sales_df['Voucher'] = sales_df['Voucher'].str.slice(start=4,stop=None,step=1).astype(int)
```

```python
# Coverting the Party and product column into uppercase
sales_df['Party'] = sales_df['Party'].str.upper()
sales_df['Product'] = sales_df['Product'].str.upper()
sales_df.head(15)
```

```python
# Converting the Qty column into an integer
sales_df['Qty'] = sales_df['Qty'].str.replace(',','').astype(float).astype(int)
```

```python
# Removing commas in the Rate column
sales_df['Rate'] = sales_df['Rate'].str.replace(',','').astype(float)
```

```python
# Sorting the Sales Transaction file in the order of Date and Voucher
sales_df = sales_df.sort_values(['Date','Voucher'])
```

```python
sales_df.head(15)
```

```python
# Then we write the Edited Sales Transaction file to .csv
sales_df.to_csv('Sales-Transactions-Edited.csv',index=False)
```

PRODUCT RANKING>>>>

```python
# Finding the no of units sold of each product
# Finding the unit price of each product (max of price considered, may required to be changed to median or mean)
top_sell_items_df = sales_df.groupby('Product').agg({'Qty':'sum', 'Rate':'max'})

```

```python
# Reset the index: converts the Product into a column
top_sell_items_df.reset_index(inplace=True)
```

```python
# Rank the product by most Qty sold
top_sell_items_df['Top_Sell_Rank'] = top_sell_items_df['Qty'].rank(method='min',ascending=False).astype(int)
top_sell_items_df.sort_values('Qty',ascending=False).head(20)
```

Most Popular Items

```python
# Considered Date column instead of Voucher, in counting the no of orders placed for a product.
# This ignores the multiple no of orders created in a single day.

# Removing duplicate records at Product, Date and Party level under the belief that
#a customer places a 2nd order of the same product in a day, only when he/she notices a wrong qty placed on the order
unique_order_items_df = sales_df.drop_duplicates(['Product','Date','Party'])

# Finding the no of orders placed and the unique no of customers placed orders, of each product
most_popular_items_df = unique_order_items_df.groupby('Product').agg({'Date':'count', 'Party':'nunique'})
most_popular_items_df.columns=['No_of_Orders','No_of_Customers']
```

```python
# Reset the index by converting the Product into a column
most_popular_items_df.reset_index(inplace=True)
most_popular_items_df.head(15)
```

```python
#finding the most frequently purchased items
O = most_popular_items_df['No_of_Orders']
C = most_popular_items_df['No_of_Customers']
M = most_popular_items_df['No_of_Customers'].max()

most_popular_items_df['Weighted_No_of_Orders'] = O * (C / M)
most_popular_items_df['Popularity_Rank'] = most_popular_items_df['Weighted_No_of_Orders'].rank(method='min',ascending=False).astype(int)

most_popular_items_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
#Merging all the Ranks that we perfomed previously
# Merge Top Selling Items Rank and Popularity Rank dataframes
product_rankings_df = pd.merge(top_sell_items_df,most_popular_items_df,how='inner',on='Product')

# Get only the Product, Price and Rank columns
product_rankings_df = product_rankings_df[['Product','Rate','Top_Sell_Rank','Popularity_Rank']]

# List the Product Rankings
product_rankings_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
#Write the Product Rankings into a .csv file
product_rankings_df.to_csv('Product-Rankings.csv',index=False)
```

```python
#Create a Pickle (.pkl) file with the Ranking dataframe
import pickle
pickle.dump(product_rankings_df, open('prod_ranking_model.pkl','wb'))
```

```python
# Import Feature Engineered Sales Transaction file
sales_df = pd.read_csv('Sales-Transactions-Edited.csv')
```

Building a Correlation Matrix for the Customer-Product relations (using User-User based recommendation)

```python
# Find the total qty purchased by each customer of each product
prod_cust_qty_df = sales_df.groupby(['Product','Party']).agg({'Qty':'sum'})

# Reset the index :converts the Party and Product into columns
prod_cust_qty_df.reset_index(inplace=True)
prod_cust_qty_df.head(15)
```

```python
# Find the no of unique customers purchased each product
prod_cust_count_df = sales_df.groupby(['Product']).agg({'Party':'nunique'})

# Set the customer count column
prod_cust_count_df.columns=['No_of_Customers']

# Reset the index :converting the Party and Product into columns
prod_cust_count_df.reset_index(inplace=True)
prod_cust_count_df.head(15)
```

```python
# Merging the unique customer count and qty purchased of each product
prod_cust_df = pd.merge(prod_cust_qty_df,prod_cust_count_df,how='inner',on='Product')
```

```python
# Creating a pivot table with all Customers on columns and Products on rows, and Qty as values
prod_cust_pivot_df = prod_cust_df.pivot(index='Product',columns='Party',values='Qty').fillna(0)
```

```python
# Identifying the correlation between every two customers and building a correlation matrix using corr() method
cust_correlation_df = prod_cust_pivot_df.corr(method='spearman',min_periods=5)
```

```python
#saving data as csv
cust_correlation_df.to_csv('Customer-Customer-Correlation-Matrix.csv')
```

```python
#Creating a Pickle (.pkl) file with the Correlation Matrix dataframe
pickle.dump(cust_correlation_df, open('cust_correlation_model.pkl','wb'))
```

```python
import matplotlib.pyplot as plt
import pickle

# Import Feature Engineered Sales Transaction file
sales_df = pd.read_csv('Sales-Transactions-Edited.csv')
```

```python
#ITERMS CUSTOMER PURCHASED THE MOST
#Find the no of units sold of each product by customer
top_sell_cust_items_df = sales_df.groupby(['Party','Product']).agg({'Qty':'sum'})

# Reset the index :converting the Party and Product into a column
top_sell_cust_items_df.reset_index(inplace=True)
```

```python
# Rank the product by most Qty sold, at Customer level
party_col = top_sell_cust_items_df['Party']
qty_col = top_sell_cust_items_df['Qty'].astype(str)
top_sell_cust_items_df['Top_Sell_Rank'] = (party_col + qty_col).rank(method='min',ascending=False).astype(int)

top_sell_cust_items_df.sort_values('Top_Sell_Rank',ascending=True).head(20)
```

```python
# Considered Date column instead of Voucher, in counting the no of orders placed for a product.
# This ignores the multiple no of orders created in a single day.
# Remove duplicate records at Party, Product and Date level
unique_order_items_df = sales_df.drop_duplicates(['Party','Product','Date'])
```

```python
# Find the no of orders placed and the unique no of customers placed orders, of each product
freq_items_df = unique_order_items_df.groupby(['Party','Product']).agg({'Date':'count'})
freq_items_df.columns=['No_of_Orders']
# Reset the index by converting the Party and Product into columns
freq_items_df.reset_index(inplace=True)
```

```python
# Rank the product by No of Orders, at Customer Level
party_col = freq_items_df['Party']
ord_count_col = freq_items_df['No_of_Orders'].astype(str)
freq_items_df['Popularity_Rank'] = (party_col + ord_count_col).rank(method='min',ascending=False).astype(int)

# List of top 20 most popular items sold
freq_items_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
# Merge Top Selling Items Rank and Popularity Rank dataframes
cust_prod_rankings_df = pd.merge(top_sell_cust_items_df,freq_items_df,how='inner',on=['Party','Product'])
```

```python
# Merge the Unit Price (max price at product level)
# Find the unit price of each product (max of price considered, may required to be changed to median or mean)
items_price_df = sales_df.groupby(['Product']).agg({'Rate':'max'})

# Reset the index by converting the Party and Product into columns
items_price_df.reset_index(inplace=True)
```

```python
# This ensures the same unit price is attached to the product purchased by different customers
cust_prod_rankings_df = pd.merge(cust_prod_rankings_df,items_price_df,how='left',on='Product')
# Get only the Customer, Product, Price and Rank columns
cust_prod_rankings_df = cust_prod_rankings_df[['Party','Product','Rate','Qty','Top_Sell_Rank','No_of_Orders','Popularity_Rank']]

# List the Product Rankings
cust_prod_rankings_df.sort_values('Popularity_Rank',ascending=True).head(20)
```

```python
cust_prod_rankings_df.to_csv('Customer-Product-Rankings.csv',index=False)
```

```python
pickle.dump(cust_prod_rankings_df, open('cust_prod_ranking_model.pkl','wb'))
```
